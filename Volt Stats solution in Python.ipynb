```python
import requests as r 
import pandas as pd 
# We don't need BeautifulSoup here because again we're just pulling from the JSON request 
# All we actually need is to request the file, get the JSON
resp = r.get("https://www.voltstats.net/Stats/LeaderboardTableJson")
txt = resp.json()
# read in the Data dictionary object into a data frame 
df = pd.DataFrame(txt['Data'])
# Give it column names. Probably not necessary but easy for dropping in the next step 
df.columns = ['BLANK', 'id_serial', 'nickname', 'state', 'ev_miles', 
              'total_miles', 'ev_perc', 'mpg', 'mpg_e', 'mpg_cs']
# Split the id column to get the id and serial columns 
df[["id", "serial"]] = df['id_serial'].str.split("~",expand = True)
# get rid of unneeded columns 
df.drop(['BLANK', "id_serial"], axis = 1, inplace = True)
df.to_csv(r'voltstats.csv', index = False, header = True)
df
```
